3. Lakukan pengukuran kecenderungan memusat berdasarkan nilai mean, median, dan modus untuk semua variabel kuantitatif.


```python
df = pd.DataFrame(data, columns=["Lama Belajar", "Nilai Ujian", "Kehadiran", "Ekstrakulikuler", "Usia"])

df_numeric = df.select_dtypes(include=[np.number])

mean_values = df_numeric.mean()
median_values = df_numeric.median()
mode_values = df_numeric.mode().iloc[0]

tendensi_memusat = pd.DataFrame({
    "Mean": mean_values,
    "Median": median_values,
    "Modus": mode_values
})

print(tendensi_memusat)
```
